```python
import traceback
def convert_date_string_to_datetime(df):
    date_columns = [
        'date_cluster_confirmed_public_health', 'date_cluster_no_longer_active',
        'earliest_reported_cases', 'latest_reported_cases', 
        'closure_date',
    ]
    try:
        df[date_columns] = df[date_columns].apply(pd.to_datetime, format='%m/%d/%Y')
    except Exception as error:
        exc_type, exc_obj, tb = sys.exc_info()
        f = tb.tb_frame
        lineno = tb.tb_lineno
        filename = f.f_code.co_filename
        df['flag'] = 1
        # problem_rows['file'] = df['file']
        # problem_rows['doc_page_number'] = df['doc_page_number']
        print(f'\tProblem with {df["file"]} page {df["doc_page_number"]} (row {df.name}: {error})')
    return df
    # return df, problem_rows


def process_df(
        df, description='pdfs_parsed_2023-04-02', save=False, 
        csv_path=None,
        pickle_path=r'C:\Users\silvh\OneDrive\lighthouse\portfolio-projects\BC-Covid-workplace-closures\data\interim'
        ):
    """
    Iterate through each row of the dataframe and convert date strings to datetime objects.
    If unable to convert to datetime object, flag the row and save the problematic rows and pages.
    """
    df2 = df.copy()
    df2['flag'] = ''
    problem_rows = pd.DataFrame()
    df2 = df2.transpose().apply(lambda x: convert_date_string_to_datetime(x)).transpose()
    problem_rows = df2[df2['flag'] == 1]
    df2 = df2[df2['flag'] != 1]
    problem_pages = problem_rows[['file', 'doc_page_number']].drop_duplicates()
    print(f'Rows with misaligned columns: {df2[df2["flag"] != ""]["flag"].count()} out of {df2.shape[0]}')
    if save:
        try:
            save_output(
                df2, description=description+'_okay_rows', pickle_path=pickle_path, csv_path=csv_path)
            save_output(
                problem_rows, description=description+'_bad_rows', pickle_path=pickle_path, csv_path=csv_path)
            save_output(
                problem_pages, description=description+'_problem_pages', pickle_path=pickle_path, csv_path=csv_path)
        except:
            print('Unable to save outputs')
    return df2, problem_rows, problem_pages
```

## `process_by_page`

```python
import re
import traceback
def convert_date_string_to_datetime_bycolumn(df):
    date_columns = [
        'date_cluster_confirmed_public_health', 'date_cluster_no_longer_active',
        'earliest_reported_cases', 'latest_reported_cases', 
        'closure_date',
    ]
    for column in date_columns:
        try:
            df[column] = pd.to_datetime(df[column], format='%m/%d/%Y')
        except Exception as error:
            exc_type, exc_obj, tb = sys.exc_info()
            f = tb.tb_frame
            lineno = tb.tb_lineno
            filename = f.f_code.co_filename
            df[column] = datetime(1900, 1, 1, 0, 0)
            print(f'\tProblem with {df["file"]} page {df["doc_page_number"]} column {column} (row {df.name}: {error})')
    return df

def split_column(df, 
        column_to_split='latest_reported_cases', 
        column1='latest_reported_cases', column2='workplace_classification'):
    split_text = df[column_to_split].split(' ')
    if split_text[0] == split_text[-1]:
        df[column1] = None
        df[column2] = split_text[-1] 
    else:
        df[column1] = split_text[0]
        df[column2] = split_text[-1]

    print(f'Column "{column_to_split}" split to {df[column1]} and {df[column2]} for columns "{column1}" and "{column2}"')
    return df

def map_values(df, column, mapping):
    df[column] = df[column].str.upper().map(mapping, na_action=None)
    if column == 'city':
        mapping = {
            'SURREY': 'SURREY',
            'EMEOS': 'CAWSTON/KEREMEOS',
            'BURNABY': 'BURNABY',
            'DELTA': 'DELTA',
            'LANGLEY': 'LANGLEY',
            'VANCOUVER': 'VANCOUVER',
        }
    return df

def process_by_page(
        df, index, function, args=None, fixed_df_dict=fixed_df_dict,
        problem_pages_df=problem_pages, 
        part_num_column = 'file', page_num_column = 'doc_page_number',
        save=False,
        ):
    file = problem_pages_df.iloc[index][part_num_column]
    page_num = problem_pages_df.iloc[index][page_num_column]
    part_num = re.search(r'Part(\d+)', file).group(0)
    print(f'Problem pages implicit index {index}: {file} page {page_num}')
    # print(f'\tDataFrame columns: {df.columns}')
    filter = (df[part_num_column] == file) & (df[page_num_column] == page_num)
    page = df[filter].copy()
    print(f'Applying function {function} to {file} page {page_num}')
    print(f'\tFunction arguments: {args}')
    args = [args] if type(args) == str else args
    if function == map_values:
        column = args[0]
        if (args[1] == True) & (type(args[1]) != dict):
        # if (type(args[1]) != dict):
            if args[0] == 'city':
                mapping = {
                    'SURREY': 'SURREY',
                    'EMEOS': 'CAWSTON/KEREMEOS',
                    'BURNABY': 'BURNABY',
                    'DELTA': 'DELTA',
                    'LANGLEY': 'LANGLEY',
                    'VANCOUVER': 'VANCOUVER',
                }
            elif args[0] == 'workplace_classification':
                mapping = {
                    'Other': 'Other',
                    'Retail': 'Retail',
                    'nstruction': 'Maintainance/Trades/Construction',
                    'Office': 'Office',
                    'ture': 'Farm/Agriculture',
                    'Processing': 'Food Processing',
                    'on': 'Transportaton',
                    'Bar/Lounge': 'Restaurant/Bar/Lounge',
                    'Warehouse': 'Warehouse',
                    'Other': 'Other',
                    'ng/Industry': 'Manufacturing/Industry',
                }
            elif args[0] == 'workplace_status':
                mapping = {
                    'transmission': 'Not closed: insufficient evidence of recent transmission',
                    'Closed': 'Closed',
                    'open': 'Not closed: Public interest in remaining open',
                    'closed': 'Not closed',
                    'measures': 'Not closed: Alternate control measures',
                    'closed)': 'Managed without closure (not closed)',
                    'PH': 'Partially closed or widespread isolation by PH',
                    'order': 'Closed by order',
                }
                print(f'mapping: {mapping}')
            print('Using default mapping...')
        else:
            mapping = args[1]
        page[column] = page[column].str.upper() if args[0] == 'city' else page[column]
        page[column] = page[column].map(mapping, na_action=None)
        processed_page = page
    elif function == 'replace':
        column = args[0]
        page[column] = page[column].replace(args[1], args[2])
        processed_page = page
    else:
        processed_page = page.apply(function, args=args, axis=1)
        if function == split_column:
            if args[0] == 'total_cases':
                map_column = 'workplace_status'
                mapping = {
                    'transmission': 'Not closed: insufficient evidence of recent transmission',
                    'Closed': 'Closed',
                    'open': 'Not closed: Public interest in remaining open',
                    'closed': 'Not closed',
                    'measures': 'Not closed: Alternate control measures',
                    'closed)': 'Managed without closure (not closed)',
                    'PH': 'Partially closed or widespread isolation by PH',
                    'order': 'Closed by order',
                    'non-closure)': 'Open (if open, reason for non-closure)'
                    }
            elif args[0] == 'latest_reported_cases':
                map_column = 'workplace_classification'
                mapping = {
                    'Other': 'Other',
                    'Retail': 'Retail',
                    'nstruction': 'Maintainance/Trades/Construction',
                    'Office': 'Office',
                    'ture': 'Farm/Agriculture',
                    'Processing': 'Food Processing',
                    'on': 'Transportaton',
                    'Bar/Lounge': 'Restaurant/Bar/Lounge',
                    'Warehouse': 'Warehouse',
                    'Other': 'Other',
                    'ng/Industry': 'Manufacturing/Industry',
                    'Gym/Fitness': 'Gym/Fitness',
                    'Care': 'Personal Care',
                    'Responder': 'First Responder',
                    '(Acute/LTC)': 'Health Care Facility (Acute/LTC)',
                }
            if args[0]!= 'days_since_last_case':
                print(f'Map_column: {[value for value in processed_page[map_column]]}')
                processed_page[map_column] = processed_page[map_column].map(mapping, na_action=None)
                print(f'Values in {map_column} mapped to {mapping}')
    if save:
        fixed_df_dict[f'{part_num}_page{page_num}'] = processed_page
        print(f'Updated fixed_df_dict keys: {list(fixed_df_dict.keys())}')
        print(f'Date and time completed: {datetime.now().strftime("%Y-%m-%d %H:%M:%S")}\n\n')
        return fixed_df_dict
    else:
        print('DataFrame columns: ',[f'{column}: {dtype}' for column, dtype in processed_page.dtypes.items()], '\n\n')
        return processed_page

```

## `view_problem_page` function

```python
def view_problem_page(index, problem_pages_df=problem_pages, problem_rows_df=problem_rows, 
    part_num_column = 'file', page_num_column = 'doc_page_number'):
    date_columns = [
        'date_cluster_confirmed_public_health', 'date_cluster_no_longer_active',
        'earliest_reported_cases', 'latest_reported_cases', 
        'closure_date',
    ]
    integer_columns = [
        'cases_when_cluster_confirmed', 
        'total_cases', 'days_since_last_case',
    ]
    file = problem_pages_df.iloc[index][part_num_column]
    page_num = problem_pages_df.iloc[index][page_num_column]
    print(f'Implicit index {index}: {file} page {page_num}')
    filter = (problem_rows_df[part_num_column] == file) & (problem_rows_df[page_num_column] == page_num)
    page = problem_rows_df[filter]
    page_copy = page.copy()
    problem_date_columns = []
    problem_integer_columns = []
    for column in date_columns:
        try:
            page_copy[column] = page_copy[column].apply(pd.to_datetime, format='%m/%d/%Y')
        except:
            problem_date_columns.append(column)
    for column in integer_columns:
        try:
            page_copy[column] = page_copy[column].apply(pd.to_numeric)
        except:
            problem_integer_columns.append(column)
    print(f'Could not convert to datetime: {[column for column in problem_date_columns]}')
    print(f'Could not convert to integer: {[column for column in problem_integer_columns]}')
    print(f'Page columns: {[column for column in page.columns]}')
    return page.drop(columns=['file', 'doc_page_number', 'flag'])#, 'cluster_confirmed_epidate', 'cluster_closed_epidate'])
```

### `view_page` function

```python
def view_page(df, part_num, page_num,
    filter_column1 = 'file', filter_column2 = 'doc_page_number'):
    date_columns = [
        'date_cluster_confirmed_public_health', 'date_cluster_no_longer_active',
        'earliest_reported_cases', 'latest_reported_cases', 
        'closure_date',
    ]
    integer_columns = [
        'cases_when_cluster_confirmed', 
        'total_cases', 'days_since_last_case',
    ]
    filter = (df[filter_column1].str.contains(f'Part{part_num}')) & (df[filter_column2] == page_num)
    page_copy = df[filter].copy()
    problem_date_columns = []
    problem_integer_columns = []
    for column in date_columns:
        try:
            page_copy[column] = page_copy[column].apply(pd.to_datetime, format='%m/%d/%Y')
        except:
            problem_date_columns.append(column)
    for column in integer_columns:
        try:
            page_copy[column] = page_copy[column].apply(pd.to_numeric)
        except:
            problem_integer_columns.append(column)
    print(f'Could not convert to datetime: {[column for column in problem_date_columns]}')
    print(f'Could not convert to integer: {[column for column in problem_integer_columns]}')
    print(f'Page columns: {[column for column in page_copy.columns]}')
    return page_copy

part_num = 2
page_num = 4

view_page(df, part_num, page_num)
```

## Iteration 1.1

```python
def processing_chain(
    index, steps_dict_list, problem_rows_df=problem_rows, 
    fixed_df_dict=fixed_df_dict, problem_pages_df=problem_pages,
    save_to_dict=False
    ):
    df = problem_rows_df.copy()
    for iteration, steps_dict in enumerate(steps_dict_list):
        if iteration == len(steps_dict_list) - 1:
            save = save_to_dict 
        else: 
            save = False
        function = steps_dict[0]
        args = steps_dict[1]
        print(f'({iteration}) Function: {function}. Arguments: {args}. Save: {save}')
        if function == 'split_column':
            fixed_page = process_by_page(
                df, index, split_column, args=args, 
                save=save
                )
        elif function == 'map_values':
            fixed_page = process_by_page(
                df, index, map_values, args=args, 
                fixed_df_dict=fixed_df_dict, problem_pages_df=problem_pages_df,
                save=save
                )
        elif function == 'convert_date_string_to_datetime_bycolumn':
            fixed_page = process_by_page(
                df, index, convert_date_string_to_datetime_bycolumn, 
                fixed_df_dict=fixed_df_dict, problem_pages_df=problem_pages_df,
                args=args, save=save
                )
        elif function == 'replace':
            fixed_page = process_by_page(
                df, index, 'replace', 
                fixed_df_dict=fixed_df_dict, problem_pages_df=problem_pages_df,
                args=args, save=save
                )
        if iteration < len(steps_dict_list) - 1:
            df = fixed_page
    view_problem_page(index, problem_rows_df=df)
    return fixed_page

index = 5
steps_dict_list = [
    ('split_column', ('latest_reported_cases', 'latest_reported_cases', 'workplace_classification')),
    ('split_column', ('total_cases', 'total_cases', 'workplace_status')),
    ('convert_date_string_to_datetime_bycolumn', None),
    ('split_column', ('days_since_last_case', 'days_since_last_case', 'active_clusters')),
    ('replace', ('city', 'COQUITLAM', 'PORT COQUITLAM')),
]
# print(steps_dict_list[0])
# save = False
save = True
fixed_page = processing_chain(index, steps_dict_list, save_to_dict=save)
```
