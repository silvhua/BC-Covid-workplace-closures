```python
def view_problem_page(index, problem_pages_df=problem_pages, problem_rows_df=problem_rows, 
    part_num_column = 'file', page_num_column = 'doc_page_number'):
    date_columns = [
        'date_cluster_confirmed_public_health', 'date_cluster_no_longer_active',
        'earliest_reported_cases', 'latest_reported_cases', 
        'closure_date',
    ]
    integer_columns = [
        'cases_when_cluster_confirmed', 
        'total_cases', 'days_since_last_case',
    ]
    file = problem_pages_df.iloc[index][part_num_column]
    page_num = problem_pages_df.iloc[index][page_num_column]
    print(f'Implicit index {index}: {file} page {page_num}')
    filter = (problem_rows_df[part_num_column] == file) & (problem_rows_df[page_num_column] == page_num)
    page = problem_rows_df[filter]
    page_copy = page.copy()
    problem_date_columns = []
    problem_integer_columns = []
    for column in date_columns:
        try:
            page_copy[column] = page_copy[column].apply(pd.to_datetime, format='%m/%d/%Y')
        except:
            problem_date_columns.append(column)
    for column in integer_columns:
        try:
            page_copy[column] = page_copy[column].apply(pd.to_numeric)
        except:
            problem_integer_columns.append(column)
    print(f'Could not convert to datetime: {[column for column in problem_date_columns]}')
    print(f'Could not convert to integer: {[column for column in problem_integer_columns]}')
    print(f'Page columns: {[column for column in page.columns]}')
    return page.drop(columns=['file', 'doc_page_number', 'flag'])#, 'cluster_confirmed_epidate', 'cluster_closed_epidate'])
```
