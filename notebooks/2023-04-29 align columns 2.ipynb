## `processing_chain`

```python
def processing_chain(
    index, steps_dict_list, problem_rows_df=problem_rows, 
    fixed_df_dict=fixed_df_dict, problem_pages_df=problem_pages,
    input_type=None, save_to_dict=False
    ):
    df = problem_rows_df.copy()
    for iteration, steps_dict in enumerate(steps_dict_list):
        if iteration == len(steps_dict_list) - 1:
            save = save_to_dict 
        else: 
            save = False
        function = steps_dict[0]
        args = steps_dict[1]
        print(f'({iteration}) Function: {function}. Arguments: {args}. Save: {save}')
        if function == 'split_column':
            fixed_page = process_by_page(
                df, index, split_column, args=args, 
                input_type=input_type, save=save
                )
        elif function == 'map_values':
            fixed_page = process_by_page(
                df, index, map_values, args=args, 
                fixed_df_dict=fixed_df_dict, problem_pages_df=problem_pages_df,
                input_type=input_type, save=save
                )
        elif function == 'convert_date_string_to_datetime_bycolumn':
            fixed_page = process_by_page(
                df, index, convert_date_string_to_datetime_bycolumn, 
                fixed_df_dict=fixed_df_dict, problem_pages_df=problem_pages_df,
                input_type=input_type, args=args, save=save
                )
        elif function == 'replace':
            fixed_page = process_by_page(
                df, index, 'replace', 
                fixed_df_dict=fixed_df_dict, problem_pages_df=problem_pages_df,
                input_type=input_type, args=args, save=save
                )
        elif function == 'fillnull':
            fixed_page = process_by_page(
                df, index, fillnull, 
                fixed_df_dict=fixed_df_dict, problem_pages_df=problem_pages_df,
                input_type=input_type, args=args, save=save
                )
        elif function == 'shift_columns': 
            fixed_page = shift_columns(df, columns_affected=args[0],
                period=args[1], replacement_values=args[2] if len(args) > 2 else None)
        if iteration < len(steps_dict_list) - 1:
            df = fixed_page
    view_problem_page(index, problem_rows_df=df)
    print(f'fixed_df_dict keys: {[key for key in list(fixed_df_dict.keys())[-1:0:-1]]}')
    return fixed_page

# index = 6
steps_dict_list = [
    ('split_column', ('latest_reported_cases', 'latest_reported_cases', 'workplace_classification')),
    ('split_column', ('total_cases', 'total_cases', 'workplace_status')),
    ('convert_date_string_to_datetime_bycolumn', None),
    ('split_column', ('days_since_last_case', 'days_since_last_case', 'active_clusters')),
    ('replace', ('city', 'COQUITLAM', 'PORT COQUITLAM')),
]
```

## `clear_and_shift_left`

```python
def clear_and_shift_left(problem_page_index, 
    columns_to_clear, columns_to_keep,
    df_index=None, rightmost_values=None):
    
    if type(df_index) == int:
        page_df = view_problem_page(problem_page_index).loc[df_index:].head(1)
    elif df_index == None:
        page_df = view_problem_page(problem_page_index)
    else:
        page_df = view_problem_page(problem_page_index).loc[df_index]
    new_df = page_df.drop(columns=columns_to_clear+columns_to_keep)
    new_df.columns = page_df.columns[:len(new_df.columns)]
    if rightmost_values:
        for column, value in rightmost_values.items():
            new_df[column] = value
    merged_df = pd.concat([new_df, page_df[columns_to_keep]], axis=1)
    return merged_df 

columns_to_clear = ['city', 'date_cluster_confirmed_public_health']
rightmost_values = {'cluster_confirmed_epidate': '17-2021'}
columns_to_keep = ['doc_page_number', 'cluster_closed_epidate', 'file', 'flag']
df_index = 2953
clear_and_shift_left(
    index, columns_to_clear, df_index=df_index,
    columns_to_keep=columns_to_keep, rightmost_values=rightmost_values)
```

```python
def align_problem_page(problem_page_index, columns_to_keep, df_index=None, value_to_ignore='s.22(1)', 
    value_to_ignore_start_index=1, value_to_ignore_end_index=None,
    blank_columns=['workplace_name', 'workplace_address']
    ):
    """
    Create dataframe from a PDF file using 
    """
    if type(df_index) == int:
        df = view_problem_page(problem_page_index).loc[df_index:].head(1)
    elif df_index == None:
        df = view_problem_page(problem_page_index)
    else:
        df = view_problem_page(problem_page_index).loc[df_index]
    print(df.index)
    print(f'\tOriginal df shape: {df.shape}')
    original_columns = df.columns
    right_columns1 = df[columns_to_keep]
    right_columns2 = df[df.columns[-len(blank_columns)-len(columns_to_keep):-len(columns_to_keep)]]
    print(f'\nRight columns 2: {right_columns2.columns}')
    
    df = df.drop(columns=columns_to_keep+right_columns2.columns.tolist())
    # Align the columns
    value_to_ignore_regex = rf'.*{re.escape(value_to_ignore[value_to_ignore_start_index:value_to_ignore_end_index])}.*'
    dropped_columns = []
    if df.iloc[0].str.contains(value_to_ignore_regex).any():
        row_values = df.fillna(0).iloc[0].values.tolist()
        for index, item in enumerate(row_values):
            if re.match(value_to_ignore_regex, str(item)):
                value_to_ignore_column = index
                break

        print(f'\tRemoving {value_to_ignore} from column {value_to_ignore_column}')
        df.iloc[0, value_to_ignore_column:-1] = df.iloc[0, value_to_ignore_column+1:].values
        
        print(f'Replacing {df.columns[-1]} with first row replacement value: {right_columns2.columns[0]}')
        print(f'\t\tintermediate columns 2: {[column for column in df.columns]}') 

    for column in blank_columns:
        try:
            if df[column].isna().sum() != len(df): 
                blank_column_index = df.columns.tolist().index(column)
                print(f'\tShifting data from column {blank_column_index}({column})')
                df.iloc[:,blank_column_index+1:] = df.iloc[:,blank_column_index:-1].values 
                df.iloc[:,blank_column_index] = None
        except:
            pass
    column_after_blank_columns_index = df.columns.get_loc(blank_columns[-1]) + 1
    column_after_blank_columns = df.columns[column_after_blank_columns_index]
    
    try: # replacing missing values
        while df[column_after_blank_columns].isna().sum() == len(df):
            print(f'\tShifting data to the right until `{column_after_blank_columns}` column')
            df.iloc[:,column_after_blank_columns_index:-1] = df.iloc[:,column_after_blank_columns_index+1:].values 
            df.iloc[:,-1] = right_columns2.iloc[:, 0].values
        right_columns2 = right_columns2.shift(periods=-(right_columns2.shape[1]-1), axis=1)
        right_columns2.iloc[:, -1] = right_columns1.iloc[:, 0].values 
    except Exception as error:
        exc_type, exc_obj, tb = sys.exc_info()
        f = tb.tb_frame
        lineno = tb.tb_lineno
        filename = f.f_code.co_filename
        print("\n** An error occurred on line", lineno, "in", filename, ":", error)
    try:
        for column in df.filter(regex='blank').columns:
            if df[column].notna().sum() == 0:
                dropped_columns.append(df[column])
                df.drop(columns=column, inplace=True)
                print(f'\tDropping blank column `{column}`')
    except:
        pass
    # # final_df =  pd.concat([df, pd.concat(dropped_columns), right_columns2, right_columns1], axis=1)
    # final_df =  pd.concat([df, right_columns2, right_columns1], axis=1)
    # final_df.columns = original_columns
    # final_df = pd.concat([df, right_columns], axis=1)
    final_df = pd.concat([df, right_columns2, right_columns1], axis=1)
    final_columns = final_df.columns
    difference = set(original_columns) - set(final_columns)
    difference2 = set(final_columns) -set(original_columns) 
    print(f'Original columns (shape = {len(df), len(original_columns)}): {[column for column in original_columns]}')
    print(f'\nFinal dataframe columns (shape = {final_df.shape}): {[column for column in final_df.columns]}')
    print(f'Missing columns: {difference}. Extra columns: {difference2}')
    return final_df[original_columns]
    # return df


# columns_to_keep = ['doc_page_number', 'cluster_closed_epidate', 'file', 'flag']

columns_to_keep2 = ['cluster_confirmed_epidate', 'doc_page_number', 'cluster_closed_epidate', 'file', 'flag']
align_problem_page(index, columns_to_keep=columns_to_keep2, df_index=view_problem_page(index).index[1:].tolist()).head()
```
