```python
def convert_date_strings_to_datetime(df):
    date_columns = [
        'date_cluster_confirmed_public_health', 'date_cluster_no_longer_active',
        'earliest_reported_cases', 'latest_reported_cases', 
        'closure_date',
        # 'cluster_closed_epidate'
    ]
    df2 = df.copy()
    df2['flag'] = ''
    for col in date_columns:
        try:
            df2[col] = pd.to_datetime(df2[col], format='%m/%d/%Y')
        except:
            df2.loc[df2[col].apply(lambda x: isinstance(x, str)), 'flag'] = col
            print('\tProblem with column: ', col)
            continue
    print(f'Rows with misaligned columns: {df2[df2["flag"] != ""]["flag"].count()} out of {df2.shape[0]}')
    print(df2[date_columns].dtypes)
    return df2

convert_date_strings_to_datetime(firstpage)
```
