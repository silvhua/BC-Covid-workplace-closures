```python
import traceback
def convert_date_string_to_datetime(df, date_columns=1):
    """
    Converts date columns in the input dataframe from string format to datetime format.
    
    Args:
    - df: pandas.DataFrame. The input dataframe.
    
    Returns:
    - pandas.DataFrame. The modified input dataframe with date columns converted to datetime format.
    """
    if (date_columns == True) & (type(date_columns) != dict):
        date_columns = [
            'date_cluster_confirmed_public_health', 'date_cluster_no_longer_active',
            'earliest_reported_cases', 'latest_reported_cases', 
            'closure_date'
        ]
    try:
        df[date_columns] = df[date_columns].apply(pd.to_datetime, format='%m/%d/%Y')
    except Exception as error:
        exc_type, exc_obj, tb = sys.exc_info()
        f = tb.tb_frame
        lineno = tb.tb_lineno
        filename = f.f_code.co_filename
        df['flag'] = 1
        print(f'\tProblem with {df["file"]} page {df["doc_page_number"]} (row {df.name}: {error})')
    return df


def process_df(
        df, description='pdfs_parsed_2023-04-30', save=False, 
        csv_path=None,
        pickle_path=r'C:\Users\silvh\OneDrive\lighthouse\portfolio-projects\BC-Covid-workplace-closures\data\interim'
        ):
    """
    Processes the input dataframe by calling the `convert_date_string_to_datetime` function and
    separating the input dataframe into two dataframes based on whether their rows have correct or incorrect
    formatting. If `save` is True, saves the output dataframes to CSV and/or pickle files.
    
    Args:
    - df: pandas.DataFrame. The input dataframe.
    - description: str. A description of the output dataframes. Default is 'pdfs_parsed_2023-04-30'.
    - save: bool. Whether to save the output dataframes to files. Default is False.
    - csv_path: str or None. The file path to save the CSV files. Default is None.
    - pickle_path: str. The file path to save the pickle files. 
    
    Returns:
    - pandas.DataFrame. The modified input dataframe with date columns converted to datetime format.
    - pandas.DataFrame. A dataframe containing rows with misaligned columns.
    - pandas.DataFrame. A dataframe containing the filenames and page numbers of the pages with misaligned columns.
    """

    df2 = df.copy()
    df2['flag'] = None
    problem_rows = pd.DataFrame()
    df2 = df2.apply(lambda x: convert_date_string_to_datetime(x), axis=1)
    problem_rows = df2[df2['flag'] == 1]
    df2 = df2[df2['flag'] != 1]
    problem_pages = problem_rows[['file', 'doc_page_number']].drop_duplicates()
    print(f'Rows with misaligned columns: {df2[df2["flag"] == 1]["flag"].count()} out of {df2.shape[0]}')
    if save:
        try:
            save_output(
                df2, description=description+'_okay_rows', pickle_path=pickle_path, csv_path=csv_path)
            save_output(
                problem_rows, description=description+'_bad_rows', pickle_path=pickle_path, csv_path=csv_path)
            save_output(
                problem_pages, description=description+'_problem_pages', pickle_path=pickle_path, csv_path=csv_path)
        except:
            print('Unable to save outputs')
    return df2, problem_rows, problem_pages

save=False
df, problem_rows, problem_pages = process_df(df, save=save)
```

```python
def clean_text(df, str_cols=None, exclude=['workplace_name', 'workplace_address']):
    """
    Converts all string columns in a pandas DataFrame to lower case, strips extra white spaces, and removes `\r` characters.

    Args:
        - df (pandas DataFrame): DataFrame containing string columns to be cleaned.

    Returns:
        pandas DataFrame: DataFrame with cleaned string columns.
    """
    str_cols = df.select_dtypes(include=['object']).columns
    str_cols = [col for col in str_cols if col not in exclude]
    print(f'Cleaning string columns: {str_cols}')
    for col in str_cols:
        print(f'\t{col}')
        df[col] = df[col].apply(
            lambda x: x.lower().strip().replace('\r', ' ') if (x !=None) & (type(x) == str) else None)
    return df


df = clean_text(df)
df
```
