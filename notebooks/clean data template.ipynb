# Functions

```python
def clean_text(df, str_cols=None, exclude=['workplace_name', 'workplace_address']):
    """
    Converts all string columns in a pandas DataFrame to lower case, strips extra white spaces, and removes `\r` characters.

    Args:
        - df (pandas DataFrame): DataFrame containing string columns to be cleaned.

    Returns:
        pandas DataFrame: DataFrame with cleaned string columns.
    """
    str_cols = df.select_dtypes(include=['object']).columns
    str_cols = [col for col in str_cols if col not in exclude]
    print(f'Cleaning string columns: {str_cols}')
    for col in str_cols:
        print(f'\t{col}')
        df[col] = df[col].apply(
            lambda x: x.lower().strip().replace('\r', ' ') if (x !=None) & (type(x) == str) else None)
    return df


df = clean_text(df)
df
```
