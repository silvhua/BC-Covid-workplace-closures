```python
from tabula import read_pdf
def parse_pdf(filenames, filepath=None, pages='all'):
    """
    Create dataframe from a PDF file using 
    """
    if type(filenames) != list:
        filenames = [filenames]
    dfs_list = []
    for filename in filenames:
        try:
            filename = f'{filepath}/'.replace('\\','/')+filename if filepath else filename
            df_list = read_pdf(filename, pages=pages)
            dfs_list.append(pd.concat(df_list))
        except:
            return pd.concat(dfs_list)
    df = pd.concat(dfs_list).reset_index(drop=True)
    df = df.set_index('Unnamed: 0')
    df.index.name = None
    return df

df = parse_pdf("F22-1445 Records_Part1.pdf", pages=[1, 2])
df
```
